### Trapezoidal Rule

At first import necessary libs

```python
import numpy as np
import matplotlib.pyplot as plt
```

Let's define a sample function to integrate

```python
def f(x):
    return 0.2 +25 * x -200 * x**2 +675 * x**3 -900 * x**4 +400 * x**5
```

Let's plot it!

```python
array = np.arange(-0.01,0.82,0.01)
plt.plot(array,f(array),label='f(x)')
plt.legend()
```

Let's define our Trapezoidal Rule!

```python
def trapezoidal_rule(f,a,b):
    return (b - a) * (f(a) + f(b)) / 2
```

```python
print(trapezoidal_rule(f,0,0.8))
plt.plot(array,f(array),label='f(x)')
plt.plot([0,0.8],[f(0),f(0.8)],color='red',label='Trapezoid_rule')
plt.legend()
```

### Simsponâ€™s 1/3 Rule
Letâ€™s implement it,

```python
def simpsons_1_3_rule(f,a,b):
    return (b-a)/6 * (f(a) + 4*f((a+b)/2) + f(b))
```

Let's plot it,

```python
from scipy.interpolate import CubicSpline

print(simpsons_1_3_rule(f,0,0.8))

plt.plot(array,f(array),label='f(x)')
plt.scatter([0,0.4,0.8],[f(0),f(0.4),f(0.8)],color='red')
plt.plot(array,CubicSpline([0,0.4,0.8],[f(0),f(0.4),f(0.8)])(array),color='green',label='Cubic Spline')
plt.fill_between(array,CubicSpline([0,0.4,0.8],[f(0),f(0.4),f(0.8)])(array),color='yellow',alpha=0.5)
plt.legend()
```
