```python
import numpy as np
import matplotlib.pyplot as plt

def linear_least_squares(x, y):
    """
    Calculates the slope and intercept for a linear regression line.
    """
    n = np.size(x)

    # Calculate mean of x and y
    mean_x, mean_y = np.mean(x), np.mean(y)

    # Calculate the cross-deviation of x and y, and the deviation about x
    SS_xy = np.sum(y * x) - n * mean_y * mean_x
    SS_xx = np.sum(x * x) - n * mean_x * mean_x

    # Calculate the regression coefficients (slope and intercept)
    b1 = SS_xy / SS_xx
    b0 = mean_y - b1 * mean_x

    return b0, b1

# 1. Generate sample data
x = np.array([5, 15, 25, 35, 45, 55])
y = np.array([5, 20, 14, 32, 22, 38])

# 2. Get the coefficients
b0, b1 = linear_least_squares(x, y)

# 3. Make predictions
y_pred = b0 + b1 * x

# 4. Plot the results
plt.figure(figsize=(8, 6))
# Plot the original data points
plt.scatter(x, y, color='blue', label='Actual data')
# Plot the regression line
plt.plot(x, y_pred, color='red', linewidth=2, label=f'Regression line: $y = {b1:.2f}x + {b0:.2f}$')

# Add labels and a title
plt.xlabel('Independent Variable (x)')
plt.ylabel('Dependent Variable (y)')
plt.title('Linear Least Squares Regression (From Scratch)')
plt.legend()
plt.grid(True)
plt.show()
```
