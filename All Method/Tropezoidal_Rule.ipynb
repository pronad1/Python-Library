### Import necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import trapezoid as trapz
```

### Define the function to integrate
For example, let's integrate $f(x)=x^2$ from a=0 to b=4

```python
def my_function(x):
    return x**2
```

### Implement the Trapezoidal Rule

```python
def trapezoidal_rule(func, a, b, n):
    delta=(b-a)/n

    # Calculate the sum of the first and last terms
    s=func(a)+ func(b)

    # Add the middle terms(multiplied by 2)
    for i in range(n):
        s += 2*func(a+i*delta)

    return (delta/2)*s
```

### Set integration parameters and calculate

```python
a=0
b=4
n=4

approx=trapezoidal_rule(my_function,a,b,n)
print(f"approximation: {approx}")
```

### Analytical solution( For comparison )

```python
exact=64/3
print(f"Exact integral: {exact}")

# Using a larger number of sub intervals for better accuracy
n_large=100
approx_large=trapezoidal_rule(my_function,a,b,n_large)
print(f"Approximate integral with n=100: {approx_large}")
```

### Using Scipy's built-in function for comparison

```python
x_values=np.linspace(a,b,n+1)
y_values=my_function(x_values)
integral_scipy=trapz(y_values,x_values)
print(f"Integral using scipy.integrate.trapz: {integral_scipy}")
```

### Visualization

```python
x=np.linspace(a,b,n+1)
y=my_function(x)

x_plot=np.linspace(a,b,100)
y_plot=my_function(x_plot)

plt.figure(figsize=(10,5))
plt.plot(x_plot,y_plot,color='blue',label='$f(x)=x^2$')

for i in range(n):
    xs=[x[i],x[i],x[i+1],x[i+1]]
    ys=[0,y[i],y[i+1],0]
    plt.fill(xs,ys,'gray',edgecolor='black',alpha=0.3)

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Tropezoidal Rule Approximation (n={n})')
plt.grid(True)
plt.show()
```
