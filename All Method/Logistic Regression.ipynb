### Logistic Regression
Logistic regression is a statistical method used for binary classification problems, when the goal is to predict the probability of an instance belonging to one of two classes. Unlike linear regression, which predicts continuous values, logistic regression predicts probabilities that are mapped to discrete classes (0 or 1).

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def logistic_regression(x):
    return 1/(1+np.exp(-x))

def predict_logistic(x,m,b):
    linear_combination = m * x + b
    return logistic_regression(linear_combination)
```

### An example usage

```python
predict_logistic(0,1,0),predict_logistic(1,1,0),predict_logistic(-1,1,0)
```

### Let's plot the curve,,,

```python
x=np.linspace(-10,10,100)

y=logistic_regression(x)
plt.plot(x,y)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification

# Generate sample data
X, y = make_classification(n_samples=100, n_features=1, n_informative=1, 
                          n_redundant=0, n_clusters_per_class=1, random_state=42)

# Create and train model
model = LogisticRegression()
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)
accuracy = model.score(X, y)

# Visualize
plt.scatter(X, y, c=y, cmap='coolwarm', label='Actual')
plt.scatter(X, y_pred, marker='x', c='green', label='Predicted')
plt.xlabel('Feature')
plt.ylabel('Class')
plt.title(f'Logistic Regression (Accuracy: {accuracy:.2f})')
plt.legend()
plt.show()

print(f"Accuracy: {accuracy:.4f}")
print(f"Coefficients: {model.coef_}")
print(f"Intercept: {model.intercept_}")
```
