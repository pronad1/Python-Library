### Interpolation
There are two interpoltion method that are Newton's Forward and Backward Interpolation methods.

### Newton's Forward Interpolation
Here is the text formatted in Markdown:Used when the point $x$ is near the beginning of the table.Formula:$$P(x) = y_0 + u \Delta y_0 + \frac{u(u-1)}{2!} \Delta^2 y_0 + \dots $$ 


```python
import numpy as np
import matplotlib.pyplot as plt

x=np.array(eval(input("Enter x points (equally spaced) : ")))
y=np.array(eval(input()))
x_val=float(input("Enter the value of x to interpolate : "))
```

```python
import math

h=x[1]-x[0]
u=(x_val - x[0])/h
```

```python
# Build forward difference table

n=len(x)
diff=np.zeros((n,n))
diff[:,0]=y
for i in range(1,n):
    for j in range(n-i):
        diff[j,i]=diff[j+1,i-1]-diff[j,i-1]

result=y[0]
u_term=1
for i in range(1,n):
    u_term*=(u - (i-1))
    result += u_term*diff[0,i]/math.factorial(i)

print(f"The interpolated value at x = {x_val} is {result}")
```

### Newtonâ€™s Backward Interpolation
Used when $x$ is near the end of the table.Formula:$$P(x) = y_n + u \Delta y_n + \frac{u(u+1)}{2!} \Delta^2 y_n + \dots$$

```python
# Assuming x, y, x_val same as above
u = (x_val - x[-1]) / h

# Build backward difference table
diff = np.zeros((n,n))
diff[:,0] = y
for i in range(1,n):
    for j in range(i,n):
        diff[j,i] = diff[j,i-1] - diff[j-1,i-1]

# Compute interpolation
result = y[-1]
u_term = 1
for i in range(1,n):
    u_term *= (u + (i-1))
    result += u_term * diff[-1,i] / math.factorial(i)

print(f"Interpolated value at x={x_val} using backward formula is {result}")
```
