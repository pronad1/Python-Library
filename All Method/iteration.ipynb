### Iteration Method
The iteration method is a common approach to find roots of equations. It involves rearranging the equation into the form x = g(x) and then iteratively applying g to an initial guess until convergence.

Let’s consider the equation x^3 - 4x + 1 = 0. We can rearrange it to x = (x^3 + 1) / 4.

```python
def f2(x):
    return x**3 - 4*x + 1
def g3(x):
    return (x**3 + 1) / 4
```

### Let’s implement the iteration methdod for this case,

```python
def iteration_method(g,x, tol=1e-4,max_iter=100):
    for i in range(max_iter):
        x_new = g3(x)
        if abs(x_new - x) < tol:
            return x_new
        x = x_new
    raise ValueError("Iteration did not converge within the maximum number of iterations.")
```

### And finally plotting time,

```python
root = iteration_method(g3,1)
# print(root)

import numpy as np
import matplotlib.pyplot as plt
x= np.arange(-10,10,0.2)

plt.plot(x,f2(x),label='f2 = x^3 - 4x + 1')
plt.scatter(root, f2(root),color='blue')
plt.axvline(root,color='red',linestyle='--',label=f'x = {root:.5f}')

plt.axvline(0, color='yellow', linestyle='--')
plt.axhline(0, color='yellow', linestyle='--')

plt.xlabel('x')
plt.ylabel('f2(x)')
plt.title(f'Iteration Method Root: {root:.5}')
plt.legend()
plt.grid(True)
```
