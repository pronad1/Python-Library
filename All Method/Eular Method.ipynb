```python
import numpy as np
import matplotlib.pyplot as plt

# Differential equation: dy/dx = x + y
def f(x, y):
    return x + y

x0 = float(input("Enter x0: "))
y0 = float(input("Enter y0: "))
h = float(input("Enter step size h: "))
xn = float(input("Enter x where you want y: "))

x_vals = [x0]
y_vals = [y0]

print("\nIteration Results:")
print("n\t x\t\t y")
print("0\t {:.4f}\t {:.4f}".format(x0, y0))

n = 0
while x0 < xn:
    y0 = y0 + h * f(x0, y0)
    x0 = x0 + h
    n += 1
    x_vals.append(x0)
    y_vals.append(y0)
    print("{}\t {:.4f}\t {:.4f}".format(n, x0, y0))

print(f"\nApproximate value at x = {xn} is y = {y0:.6f}")

# Plotting
plt.plot(x_vals, y_vals, label="Euler's Method")
plt.scatter(x_vals, y_vals)
plt.title("Euler's Method for dy/dx = x + y")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```
